```python
import pandas as pd
import numpy as np
import xarray as xr
import seaborn as sns
from alias_period import tidal_aliasing
%pylab inline
```

```python

constituent = ["2N2","EP2","J1","L2","M3","M4","MF","MI2","MKS","MM","MN4","MS4","MSF","MSQ"
            ,"MTM","N4","NI2","R2","S1","SA","SSA","T2",'M2','S2','N2','K2','K1','O1','P1','Q1']

catfile = "data/catHW95.dat"

tt = tidal_aliasing(constituent,catfile,'jason-1',output='array')
```

```python
Rc_j = np.reshape((np.repeat(tt,len(tt))), (len(tt),len(tt))).T
```

```python
Ray_J = np.zeros((np.shape(Rc_j)))
```

```python
for k in range(len(Ray_J[0])):
    Ray_J[:,k] = 1/np.abs(1/Rc_j[:,k] - 1/Rc_j[k])
```

```python
Ray_J = Ray_J.round(decimals=0)
```
